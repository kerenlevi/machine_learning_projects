```python
# Converting the columns type for further analysis 

Gender_camp['Conversions'] = Gender_camp['Conversions'].str.replace(',' , '').astype('float64')
Gender_camp['Cost'] = Gender_camp['Cost'].astype('float64')
Gender_camp['Cost / conv.'] = Gender_camp['Cost / conv.'].astype('float64')
Gender_camp['Conv. rate'] = Gender_camp['Conv. rate'].astype('str').str.rstrip('%').astype('float64')
Gender_camp['Impr.'] = Gender_camp['Impr.'].astype('str').replace(",", "")
Gender_camp['Avg. cost'] = Gender_camp['Avg. cost'].replace(' --', np.nan).astype('float64')
Gender_camp['Interaction rate'] = Gender_camp['Interaction rate'].replace(' --', np.nan)
Gender_camp['Interaction rate'] = Gender_camp['Interaction rate'].astype('str').str.rstrip('%').astype('float64')
Gender_camp['Bid adj.'] = Gender_camp['Bid adj.'].replace(' --', np.nan)
Gender_camp['Bid adj.'] = Gender_camp['Bid adj.'].astype('str').str.rstrip('%').astype('float64')
Gender_camp['Impr.'] = Gender_camp['Impr.'].str.replace(',' , '').astype('float64')
Gender_camp['Interactions'] = Gender_camp['Interactions'].str.replace(',' , '').astype('float64')
```

```python
Age_camp['Conversions'] = Age_camp['Conversions'].astype('float64')
Age_camp['Cost'] = Age_camp['Cost'].astype('float64')
Age_camp['Cost / conv.'] = Age_camp['Cost / conv.'].astype('float64')
Age_camp['Conv. rate'] = Age_camp['Conv. rate'].astype('str').str.rstrip('%').astype('float64')
Age_camp['Avg. cost'] = Age_camp['Avg. cost'].replace(' --', np.nan).astype('float64')
Age_camp['Bid adj.'] = Age_camp['Bid adj.'].replace(' --', np.nan)
Age_camp['Bid adj.'] = Age_camp['Bid adj.'].astype('str').str.rstrip('%').astype('float64')
Age_camp['Impr.'] = Age_camp['Impr.'].str.replace(',' , '').astype('float64')
Age_camp['Week']= pd.to_datetime(Age_camp['Week']).dt.strftime('%Y-%m')
Gender_camp['Week']= pd.to_datetime(Gender_camp['Week']).dt.strftime('%Y-%m')
```

```python
# Copied from preformance analysis - the following is analysis of top profitable campaigns including plotting

top_campaign = ['GM - REM1 - NA #QA V2',
 'Smart Global - TOP COUNTRY  1 #QA WL V1',
 'Discovery - REM - Global #QA V1 RG1',
 'Global - Branded #QA',
 'Discovery - REM - Global #QA V1 RG2']

top_gender = Gender_camp[Gender_camp['Campaign'].isin(top_campaign)]
top_age =  Age_camp[Age_camp['Campaign'].isin(top_campaign)]
```

```python
top_Age_pivot = pd.pivot_table(data =top_age,  index = ['Campaign','Week'], columns= 'Age',
                           values = ['Cost', 'Impr.', 'Conversions'],
                          aggfunc={'Cost': np.sum, 'Impr.' : np.sum, 'Conversions': np.sum})

top_Gender_pivot = pd.pivot_table(data = top_gender,  index = ['Campaign','Week'], columns='Gender',
                              values = ['Impr.', 'Interactions','Cost', 'Conversions'],
                             aggfunc=np.sum)
```

```python
top_Age_pivot = pd.concat([d.append(d.sum().rename((k, 'Summary'))) for k, d in top_Age_pivot.groupby(level=0)])
top_Gender_pivot = pd.concat([d.append(d.sum().rename((k, 'Summary'))) for k, d in top_Gender_pivot.groupby(level=0)])
```

```python
age_top_summary = top_Age_pivot.loc[(slice(None), 'Summary'), :]
gender_top_summary = top_Gender_pivot.loc[(slice(None), 'Summary'), :]
```
